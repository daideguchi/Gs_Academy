```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
pd.core.common.is_list_like = pd.api.types.is_list_like
```

```python
start = "2019-06-01"
end = "2021-12-31"

df = data.DataReader("^N225","yahoo",start,end)
```

```python
df.head(10)
```

```python
df.index
```

```python
date=df.index
price=df["Adj Close"]
```

```python
plt.plot(date,price)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label="Nikkei225")
plt.legend()
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label="Nikkei225")
plt.title('N225',color='blue',backgroundcolor='white',size=40,loc='center')
plt.legend()
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label="Nikkei225")
plt.title("N225",color="blue",backgroundcolor="white",size=40,loc="center")
plt.xlabel("date",color="black",size=30)
plt.ylabel("price",color="black",size=30)
plt.legend()
```

```python
span01=5
span02=25
span03=50

df["sma01"]=price.rolling(window=span01).mean()
df["sma02"]=price.rolling(window=span02).mean()
df["sma03"]=price.rolling(window=span03).mean()
```

```python
plt.figure(figsize=(30,10))
plt.subplot(2,1,1)

plt.plot(date,price,label="close",color="#99b898")
# plt.plot(date,price,label="Nikkei225")

plt.plot(date,df["sma01"],label="sma01",color="#e84a5f")
plt.plot(date,df["sma02"],label="sma02",color="#ff847c")
plt.plot(date,df["sma03"],label="sma03",color="#feceab")

plt.title("N225",color="white",backgroundcolor="grey",size=30,loc="center")
plt.xlabel("date",color="black",size=30)
plt.ylabel("price",color="black",size=30)

plt.legend()
```

```python
#リクルートHD
date=df.index
price=df["Close"]

span01=5
span02=25
span03=50

df["sma01"]=price.rolling(window=span01).mean()
df["sma02"]=price.rolling(window=span02).mean()
df["sma03"]=price.rolling(window=span03).mean()

plt.figure(figsize=(30,10))
plt.subplot(2,1,1)

plt.plot(date,price,label="close",color="#99b898")
plt.plot(date,df["sma01"],label="sma01",color="#e84a5f")
plt.plot(date,df["sma02"],label="sma02",color="#ff847c")
plt.plot(date,df["sma03"],label="sma03",color="#feceab")
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df["Volume"],label="volume",color="grey")
plt.legend()
```

```python
start = "2019-06-01"
end = "2020-05-01"
company_code = "6502.JP"
```

```python
#ファーストリテ
df[(df.index>=start) & (df.index<=end)]
df = data.DataReader(company_code, "stooq")

date=df.index
price=df["Close"]

span01=5
span02=25
span03=50

df["sma01"]=price.rolling(window=span01).mean()
df["sma02"]=price.rolling(window=span02).mean()
df["sma03"]=price.rolling(window=span03).mean()

plt.figure(figsize=(30,10))
plt.subplot(2,1,1)

plt.plot(date,price,label="close",color="#99b898")
plt.plot(date,df["sma01"],label="sma01",color="#e84a5f")
plt.plot(date,df["sma02"],label="sma02",color="#ff847c")
plt.plot(date,df["sma03"],label="sma03",color="#feceab")
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df["Volume"],label="volume",color="grey")
plt.legend()
```

```python
def company_stock(start,end,company_code):
    df[(df.index>=start) & (df.index<=end)]
df = data.DataReader(company_code, "stooq")

date=df.index
price=df["Close"]

span01=5
span02=25
span03=50

df["sma01"]=price.rolling(window=span01).mean()
df["sma02"]=price.rolling(window=span02).mean()
df["sma03"]=price.rolling(window=span03).mean()

plt.figure(figsize=(30,10))
plt.subplot(2,1,1)

plt.plot(date,price,label="close",color="#99b898")
plt.plot(date,df["sma01"],label="sma01",color="#e84a5f")
plt.plot(date,df["sma02"],label="sma02",color="#ff847c")
plt.plot(date,df["sma03"],label="sma03",color="#feceab")
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df["Volume"],label="volume",color="grey")
plt.legend()
```
